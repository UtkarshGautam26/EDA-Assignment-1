```python
def x(cols):
    list_days = []
    for j in range(0,len(cols)):
        if "DAYS_" in cols[j]:
            list_days.append(cols[j])
    return list_days
        
x(df.columns) # function to get the list columns who name consist of the word "DAYS_"

```

```python
df[x(df.columns)] = abs(df[x(df.columns)])
df.DAYS_BIRTH
df.DAYS_EMPLOYED #successfully converted negative values into positive.
```
